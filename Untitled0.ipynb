```python
#                                                              2023535
#                                                             MEAN CODE
def calculate_mean(num):
    mean = sum(num) / len(num)
    return mean
num = [1, 2, 5, 4, 5]

mean = calculate_mean(num)
print("Mean of your numbers is: ", mean)
```

```python
#                                                              2023535
#                                                             MODE CODE
from collections import Counter

def calculate_mode(num):
    frequency = Counter(num)
    mode = max(frequency, key=frequency.get)
    return mode

num = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

mode = calculate_mode(num)
print("Mode of your numbers ", mode)
```

```python
#                                                              2023535
#                                                            MEDIAN CODE
def calculate_median(num):
    num.sort()
    n = len(num)
    if n % 2 == 0:
        median = (num[n//2 - 1] + num[n//2]) / 2
    else:
        median = num[n//2]
    return median

num = [2, 3, 4, 5]
median = calculate_median(num)
print("Median of your number is ", median)
```

```python
#                                                              2023535
#                                                           VARIANCE CODE
def calculate_variance(num):
    n = len(num)
    mean = sum(num) / n
    variance = sum((x - mean) ** 2 for x in num) / n
    return variance
numbers = [5,6,7,2,3,9]
variance = calculate_variance(num)
print("Variance:", variance)
```

```python
#                                                              2023535
#                                                            RANGE CODE
def calculate_range(num):
    range_val = max(num) - min(num)
    return range_val

numbers = [6,4,5,2,3,5]
range_val = calculate_range(num)
print("Range for your number is :", range_val)
```

```python
#                                                              2023535
#                                                         STANDARD DEVIATION CODE
import math

def calculate_standard_deviation(num):
    n = len(num)
    mean = sum(num) / n
    variance = sum((x - mean) ** 2 for x in num) / n
    standard_deviation = math.sqrt(variance)
    return standard_deviation

numbers = [1, 2, 3, 4, 5]

std_deviation = calculate_standard_deviation(num)
print("Standard Deviation of given data is :", std_deviation)
```

```python
#                                                              2023535
#                                                             RANDOM CODE
import random

def generate_random_numbers(n1, start, end):
    random_numbers =[random.randint(start, end) for _ in range(n1)]
    return random_numbers
n1 = 10
start = 2
end = 20

random_numbers = generate_random_numbers(n1, start, end)
print("Random Numbers:", random_numbers)
```

```python
#                                                              2023535
#                                                           COVARIANCE CODE
def calculate_covariance(d1, d2):
    n = len(d1)
    mean1 = sum(d1) / n
    mean2 = sum(d2) / n
    covariance = sum((d1[i] - mean1) * (d2[i] - mean2) for i in range(n)) / n
    return covariance

d1 = [2,4,5,6,7,8]
d2 = [25,3,45,67,8,8]
covariance = calculate_covariance(d1, d2)
print("Covariance:", covariance)
```

```python
#                                           2023535
#                                        PIE CHART CODE
import matplotlib.pyplot as plt

labels = ['A','B','C','D']
sizes = [20,32,24,22]
colors = ['gray', 'yellowgreen', 'skyblue', 'lightcoral']
explode = (0.3,0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')

plt.title("EXAMPLE OF PIE CHART")

plt.show()
```
